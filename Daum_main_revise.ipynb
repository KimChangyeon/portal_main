```python
import json
import pandas as pd
from collections import defaultdict
from datetime import datetime
import os
from tqdm.notebook import tqdm
```

```python
naver_path = '../news data/articles/all_daily_articles_naver/'
daum_path = '../news data/articles/all_daily_articles_daum/'
naver_main_path = '../news data/main/main_naver/main/'
daum_main_path = '../news data/main/main_daum/'
naver_list = os.listdir(naver_path)
daum_list = os.listdir(daum_path)
naver_main_list = os.listdir(naver_main_path)
daum_main_list = list(map(lambda x: x[3:], os.listdir(daum_main_path+'SP/')))


print('NAVER FROM {} to {}'.format(naver_list[0], naver_list[-1]))
print('DAUM FROM {} to {}'.format(daum_list[0], daum_list[-1]))

print('NAVER MAIN FROM {} to {}'.format(naver_main_list[0], naver_main_list[-1]))
print('DAUM MAIN FROM {} to {}'.format(daum_main_list[0], daum_main_list[-1]))
```

```python
day = daum_main_list[0]
daum_file_news = 'NE_' + day
with open(daum_main_path + 'NE/' + daum_file_news, 'r') as f:
    daum_main_info = json.load(f)
daum_file_enter = 'EN_' + day
with open(daum_main_path + 'EN/' + daum_file_enter, 'r') as f:
    daum_main_enter = json.load(f)
daum_main_info.extend(daum_main_enter)
daum_file_sports = 'SP_' + day
with open(daum_main_path + 'SP/' + daum_file_sports, 'r') as f:
    daum_main_sports = json.load(f)
daum_main_info.extend(daum_main_sports)
```

```python
import copy
def extract_aid(info):
    return info['url'].split('/')[-1]

def time_period (dates):
    info = copy.deepcopy(dates[0])
    info['exposed_first'] = datetime.strftime(info['date'], '%Y%m%d')
    info['exposed_last'] = datetime.strftime(dates[-1]['date'], '%Y%m%d')
    del info['date']
    return info 
```

```python
revised_main = defaultdict(list)
for day in tqdm(daum_main_list):
    daum_file_news = 'NE_' + day
    with open(daum_main_path + 'NE/' + daum_file_news, 'r') as f:
        daum_main_info = json.load(f)
    daum_file_enter = 'EN_' + day
    with open(daum_main_path + 'EN/' + daum_file_enter, 'r') as f:
        daum_main_enter = json.load(f)
    daum_main_info.extend(daum_main_enter)
    daum_file_sports = 'SP_' + day
    with open(daum_main_path + 'SP/' + daum_file_sports, 'r') as f:
        daum_main_sports = json.load(f)
    daum_main_info.extend(daum_main_sports)

    for info in daum_main_info:
        aid = extract_aid(info)
        info['date'] = datetime.strptime(day[:8], '%Y%m%d')
        revised_main[aid].append(info)
    
revised_main = {k:v for k,v in sorted(revised_main.items(), key = lambda x: len(x[1]), reverse = True)}
```

```python
dup_path = '../news data/articles/20200310_duplicated/'
dup_list = os.listdir(dup_path)[1:]
print('DUP FROM {} to {}'.format(dup_list[0], dup_list[-1]))
```

```python
def mainDate_string (aid):
    info = g.get(aid)
    if not info:
        return None
    ef = info['exposed_first']
    el = info['exposed_last']
    if ef == el:
        return ef[:4]+'-'+ef[4:6]+'-'+ef[6:]
    else:
        return ef[:4]+'-'+ef[4:6]+'-'+ef[6:]+'~'+el[:4]+'-'+el[4:6]+'-'+el[6:]
```

```python
for day in tqdm(dup_list):
    df = pd.read_csv(dup_path + day, dtype = object)
    df['daum_mainDate'] = df.apply(lambda x: mainDate_string(x['daum_id']), axis = 1)
    df.to_csv('../news data/articles/20200316_duplicated/' + day, index = False)
```
